```python
import pandas as pd
from amm.config import PROCESSED_DATA_DIR

df = pd.read_csv(PROCESSED_DATA_DIR / 'uniswap_v3.csv')

df.head()
```

```python
import statsmodels.formula.api as smf

df['fee_tier'] = df['fee_tier'].astype(str)
df.dropna(inplace=True)
df = df[(df['time'] >= '2023-09-01') & (df['time'] <= '2023-11-31')]
model = smf.ols('market_share ~ volume + fee_tier + fee_switch + fee_tier:fee_switch + liquidity + open + high + low + close + C(chain) + C(token_pair)', data=df)
result = model.fit()
print(result.summary())
```
